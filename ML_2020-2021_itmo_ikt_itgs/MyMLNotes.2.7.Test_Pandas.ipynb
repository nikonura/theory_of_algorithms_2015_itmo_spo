###  Описание данных
#### Таблица ```transactions.csv```
##### Описание
Таблица содержит историю транзакций клиентов банка за один год и три месяца.

##### Формат данных

```
customer_id,tr_datetime,mcc_code,tr_type,amount,term_id
111111,15 01:40:52,1111,1000,-5224,111111
111112,15 15:18:32,3333,2000,-100,11122233
...
```
##### Описание полей

 - ```customer_id``` — идентификатор клиента;
 - ```tr_datetime``` — день и время совершения транзакции (дни нумеруются с начала данных);
 - ```mcc_code``` — mcc-код транзакции;
 - ```tr_type``` — тип транзакции;
 - ```amount``` — сумма транзакции в условных единицах со знаком; ```+``` — начисление средств клиенту (приходная транзакция), ```-``` — списание средств (расходная транзакция);
 - ```term_id``` — идентификатор терминала;


#### Таблица ```gender_train.csv```

##### Описание
Данная таблица содержит информацию по полу для части клиентов, для которых он известен. Для остальных клиентов пол неизвестен.

##### Формат данных
```
customer_id,gender
111111,0
111112,1
...
```

##### Описание полей
 - ```customer_id``` — идентификатор клиента;
 - ```gender``` — пол клиента;

### Таблица ```tr_mcc_codes.csv```

##### Описание
Данная таблица содержит описание mcc-кодов транзакций.

##### Формат данных
```
mcc_code;mcc_description
1000;словесное описание mcc-кода 1000
2000;словесное описание mcc-кода 2000
...
```

##### Описание полей
 - ```mcc_code``` – mcc-код транзакции;
 - ```mcc_description``` — описание mcc-кода транзакции.

#### Таблица ```tr_types.csv```

Данная таблица содержит описание типов транзакций.

```
tr_type;tr_description

...
```

````python
# Вариант 1
# переназначение df
df = transactions[['customer_id', 'amount']].copy()
df['receipt'] = df[df.amount >=0].amount
df['expenditure'] = df[df.amount <0].amount
df = df.groupby(by='customer_id', sort=True).sum()
# возможно ли выше написанный результат получить функцией в одну строку?
pd.DataFrame({'customer_id': [df.receipt.idxmax(), df.expenditure.idxmin()],
              'amount': [df.receipt.max(), df.expenditure.min()]},
             index = ['receipt', 'expenditure']
             )
````

````python
# Вариант 2 объединяем группированные по пользователю столбцы прихода, расхода, разности
# на выходе максимумы и минимумы соответсвенно каждого

df = pd.merge(transactions[transactions.amount >=0].groupby(by='customer_id', sort=True).sum(), 
         transactions[transactions.amount <0].groupby(by='customer_id', sort=True).sum(), 
         on='customer_id',
         suffixes=('_receipt', '_expenditure'))[['amount_receipt', 'amount_expenditure']]
#df['diff'] = df.amount_receipt + df.amount_expenditure
#df.idxmax()
df['diff'] = df.amount_receipt - np.abs(df.amount_expenditure)
print(df.max(), '\n', df.min())
````

````python
# представим те же значения в виде датафрейма
pd.DataFrame({'customer_id': [df.amount_receipt.idxmax(), df.amount_expenditure.idxmin(), df['diff'].idxmax(), df['diff'].idxmin()],
              'amount': [df.amount_receipt.max(), df.amount_expenditure.min(), df['diff'].max(), df['diff'].min()]},
             index = ['receipt', 'expenditure', 'diff_max', 'diff_min']
             )
````
